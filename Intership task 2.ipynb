```python
import pandas as pd
import numpy as np 
import datetime as dt
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("Online Retail.xlsx")
df.head()
```

```python
df.dropna(subset=['CustomerID'], inplace=True)
df = df[df['Quantity']>0] 
df['TotalPrice']=df['Quantity'] * df['UnitPrice']
```

```python
import datetime as dt
now = df['InvoiceDate'].max() + dt.timedelta(days=1)

rfm = df.groupby('CustomerID').agg({
    'InvoiceDate':lambda x: (now -x.max()).days, #Recency
    'InvoiceNo' : 'count', # Frequency
    'TotalPrice': 'sum' ## Monetary
})
rfm.rename(columns={'InvoiceDate':'Recency',
                    'InvoiceNo':'Frequency',
                    'TotalPrice':'Monetary'}, inplace=True)

```

```python
rfm['R_Score']= pd.qcut(rfm['Recency'],4,labels=[4,3,2,1])
rfm['F_Score']= pd.qcut(rfm['Frequency'].rank(method="first"),4,labels=[1,2,3,4])
rfm['M_Score']= pd.qcut(rfm['Monetary'],4,labels=[1,2,3,4])

rfm['RFM_Segment']=(rfm['R_Score'].astype(str)+
                    rfm['F_Score'].astype(str)+
                    rfm['M_Score'].astype(str))
rfm['RFM_Score']= rfm[['R_Score','F_Score','M_Score']].sum(axis=1)

```

```python
sns.scatterplot(x='Recency', y= 'Frequency', size ='Frequency', data=rfm)
```

```python
def rfm_segment(score):
    if score >= 9:
        return "Champions"
    elif score >= 8:
        return "Loyal Customers"
    elif score >= 7:
        return "Potential Loyalists"
    elif score >= 5:
        return "Needs Attention"
    elif score >= 3:
        return "At Risk"
    else:
        return "Hibernating"

rfm['Segment'] = rfm['RFM_Score'].apply(rfm_segment)
```

```python
segment_counts= rfm['Segment'].value_counts()
```

```python
plt.figure(figsize=(8,6))
segment_counts.plot(kind='bar')
plt.title("Customer Segments Distribution (RFM Analysis)")
plt.xlabel("Customer Segment")
plt.ylabel("Number of Customers")
plt.xticks(rotation=45)
plt.show()
```

## Marketing Ideas Suggestion

1.For "Champions", Offer VIP perks that gives early access to new products. Other than that provision of exclusive discounts and refferal program can be a good marketing strategy.

2.For 'Loyal Customers', introducing loyalty program with features like points, rewards, free gifts. Personlized recommendation based upon past purchases with addition to related items is a great marketing strategy.

3.For 'Potential Loyalists', introducing special discounts or bundles motivating to repeat purchase is a good marketing strategy

4.For 'At Risk', introducing strong discounts like next purchase will give 20% discount or personilized emails motivating to repeat purchase is a good marketing strategy

5.For 'Needs Attention', introducing Festive/seasonal promotions or Targeted re-engagement campaigns like limited-time offers motivating to repeat purchase is a good marketing strategy.

6.For 'Hibernating', using Low-cost re-engagement like broad promotions, newsletter or social media campaign like retargeting ads motivating to purchase.
